```python
x2 = list(x1)
y2 = list(y1)
x3 = []
y3 = []
for i in x2:
    res = []
    for j in i:
        res.append(int(j))
    x3.append(res)
for i in y2:
    res = []
    for j in i:
        res.append(int(j))
    y3.append(res)
```

```python
def f(x, y):
    res = []
    for i in range(len(x)):
        row = []
        for j in range(len(y)):
            sum_row = 0
            for k in range(len(x[0])):
                sum_row += x[i][k] * y[k][j]
            row.append(sum_row)
        res.append(row)
    return res
```

```python
%timeit f(x3, y3)
```

```python
def make_df(cols, ind):
    "一个简单的DataFrame"
    data = {c: [str(c) + str(i) for i in ind] for c in cols}
    return pd.DataFrame(data, ind)

make_df("ABC", range(3))
```

注意：unstack()和stack()区别

1.stack()

    功能：将列索引转换为行索引。如果是多级索引，就转换最内层的索引
    
    用途：将宽格式数据转换为长格式。适用于当你希望将 DataFrame 的列“堆叠”成一列，以便进行进一步分析或可视化。
    
2.unstack()

    功能：将行索引转换为列索引。如果是多级索引，就转换最内层的索引
    
    用途：将长格式数据转换为宽格式。适用于当你希望将 DataFrame 的行“展开”成列，以便更方便地查看数据。
    
3.关键区别

    方向：

    stack() 是将列转换为行，增加行的数量。
    
    unstack() 是将行转换为列，增加列的数量。
    
    用途：

    stack() 更常用于将宽格式数据转换为长格式，适合需要进行分组或聚合操作的场景。
    
    unstack() 更常用于将长格式数据转换为宽格式，适合于数据可视化和对比的场景。

### （4）高性能的Pandas：eval()

```python
df1, df2, df3, df4 = (pd.DataFrame(np.random.random((10000, 100))) for i in range(4))
```

```python
%timeit (df1 + df2) / (df3 + df4)
```

- 减少了复合代数式计算中间过程的内存分配

```python
%timeit pd.eval("(df1 + df2) / (df3 + df4)")
```

```python
np.allclose((df1 + df2) / (df3 + df4), pd.eval("(df1 + df2) / (df3 + df4)"))    # 验证结果是否相等
```

- 实现列间运算

```python
df = pd.DataFrame(np.random.random((1000, 3)), columns = list("ABC"))
df.head()
```

```python
res_1 = pd.eval("(df.A + df.B) / (df.C - 1)")
```

```python
res_2 = df.eval("(A + B) / (C - 1)")
```

### （4）高性能的Pandas：query()

```python
%timeit res_1 = df[(df.A < 0.5) & (df.B > 0.5)]
```

```python
%timeit res_2 = df.query("(A < 0.5) & (B > 0.5)")
```

```python
df.query("(A < 0.5) & (B > 0.5)").head()
```

```python
np.allclose(res_1, res_2)
```
